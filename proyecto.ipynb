
### 3️⃣ Agregación de Datos

Se resumen las ventas agrupando por **categoría de producto**, permitiendo analizar:
- Ingresos totales
- Cantidad vendida por categoría

Este paso facilita la identificación de patrones y categorías más relevantes.


```python
# ingreso por producto:

# Se crea columna nueva y se calcula total por venta:
df_ventas['total_venta'] = df_ventas['precio'].str.replace('$', '', regex=False).astype(float) * df_ventas['cantidad']

df_ventas.head()

#se crea la pivot table
ingreso_por_producto = pd.pivot_table(
    df_ventas,
    values = ['total_venta'],
    index = ['categoria'],
    aggfunc=sum
)

ingreso_por_producto.head()

#cantidad por categoria:
cantidad_por_categoria = pd.pivot_table(
    df_ventas,
    values = ['cantidad'],
    index = ['categoria'],
    aggfunc = 'sum'
)

cantidad_por_categoria.head()
```


### 4️⃣ Integración de Datos

Se combinan los conjuntos de datos de **ventas** y **marketing** utilizando claves comunes, con el objetivo de obtener una visión más amplia de las tendencias y el impacto de las acciones comerciales sobre las ventas.


```python
#df_marketing.head()

#primero agrupo ventas:

ventas_resumidas = df_ventas.pivot_table(
    values='total_venta',
    index='producto',
    aggfunc='sum'
).reset_index()


df_combinado = pd.merge(
    ventas_resumidas,
    df_marketing,
    on=['producto'],
    how='inner'
)

#df_combinado['roi'] = df_combinado['total_venta'] / df_combinado['costo']

df_combinado.head()

```
