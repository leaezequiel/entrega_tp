
### 2️⃣ Cálculo de ventas mensuales (script básico)

Se desarrolla un script simple en Python que permita calcular el total de ventas mensuales utilizando **variables**, **operadores aritméticos** y estructuras básicas del lenguaje.


```python
#print(df_ventas)

#como era tipo string el precio tuve que reemplazar el $ y castearlo como float:
df_ventas['precio'] = (
    df_ventas['precio']
    .replace('[\$,]', '', regex=True)
    .astype(float)
)

#calcule el total facturado de cada venta
df_ventas['total_facturado'] = (
    df_ventas['precio'] * df_ventas['cantidad']
)


df_ventas.head()

#converti la columna en formato fecha E INCLUI el dia adelante TRUE!
df_ventas['fecha_venta'] = pd.to_datetime(
    df_ventas['fecha_venta'],
    dayfirst=True
)
#Chequeo ok
#df_ventas.dtypes

#creo columna nueva s/ mes
#.dt → accede a propiedades de fechas
#.to_period('M') → convierte la fecha a un período mensual (YYYY-MM)

df_ventas['mes'] = df_ventas['fecha_venta'].dt.to_period('M')

'''
groupby('mes') → agrupa todas las ventas por mes
['total_facturado'] → selecciona la columna a analizar
.sum() → suma el total facturado de cada mes
.reset_index() → devuelve un DataFrame plano (más fácil de usar)
'''
ventas_por_mes = (
    df_ventas
    .groupby('mes')['total_facturado']
    .sum()
    .reset_index()
)

print(ventas_por_mes)
```
