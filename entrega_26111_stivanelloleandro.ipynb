### 1️⃣ Carga de datos en Google Colab

Se crea el entorno de trabajo en Google Colaboratory y se cargan los distintos sets de datos desde Google Drive o enlaces externos, almacenándolos en **DataFrames** para su análisis.



```python
import pandas as pd

def load_gdrive_csv(file_id, **kwargs):
    url = f"https://drive.google.com/uc?id={file_id}"
    return pd.read_csv(url, **kwargs)

df_ventas = load_gdrive_csv("1jlo-TjeDEkIeHrbcdSVXO8pqzTTfyE6N")
df_clientes = load_gdrive_csv("11-FPOE9X1AvxGE6AUXTCspVQJzrLT_2f")
df_marketing = load_gdrive_csv("1ZSfAcNDyo2eiioc7za_h57lvlm2RtScu")

'''
print(df_ventas)
print(df_clientes)
print(df_marketing)
'''
```


### 4️⃣ Introducción a Pandas – Análisis Exploratorio

Se realiza un análisis exploratorio inicial de los DataFrames, utilizando funciones básicas de Pandas para comprender la estructura y el contenido de los datos.


```python
#hice un diccionario con cada df para hacerlo una sola vez y ejecutar a todos:
data_frames = {
    "ventas": df_ventas,
    "clientes": df_clientes,
    "marketing": df_marketing
}

#recorri cada clave, y cada valor del diccionario ejecutando las funciones
for nombre, df in data_frames.items():
    print(f"\n--- Análisis de {nombre} ---")
    df.info()
    print(df.describe())
    #print(df.head())
    print("Shape:", df.shape)
    print("Columnas:", df.columns)
```


### 5️⃣ Calidad de Datos

Se analizan los conjuntos de datos para identificar:

- Valores nulos  
- Registros duplicados  

Se documenta el **estado inicial de los datos**, lo que permitirá tomar decisiones sobre limpieza y transformación en etapas posteriores.


```python
for nombre, df in data_frames.items():
    print(f"\n--- Análisis de {nombre} ---")
    print("Nulos:\n", df.isnull().sum())
    print("Duplicados:\n", df.duplicated().sum())
```
